```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import seaborn as sns
plt.style.use('seaborn')
```

```python
df_iris = pd.read_csv('./datasets/iris.data', names=['sepal_length', 
                                                     'sepal_width', 
                                                     'petal_length',
                                                     'petal_width',
                                                     'species'])
df_iris.head()
```

```python
print('Total rows :', df_iris.shape[0])
print('Total columns :', df_iris.shape[1])
```

```python
df_iris.isnull()
```

```python
df_iris['species'].hist()
plt.show()
```

```python
df_iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
df_iris.plot(kind='scatter', x='petal_length', y='petal_width')
plt.show()
```

```python
sns.distplot(df_iris['sepal_length'])
plt.show()
```

```python
sns.boxplot(data=df_iris)
plt.xlabel('Features')
plt.ylabel('Length in cm')
plt.show()
```

```python
sns.boxplot(data=df_iris, x='species', y='petal_length')
plt.show()
```

```python
sns.pairplot(data=df_iris, hue='species')
plt.show()
```

```python
sns.heatmap(data=df_iris.corr(), annot=True)
plt.show()
```
