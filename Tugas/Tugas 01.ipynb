```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
```

```python
df_titanic['Sex'].hist()
plt.show()
```

```python
df_titanic['Fare'].hist()
plt.show()
```
