```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns

plt.style.use('seaborn')
```

```python
df_ytb_all = pd.read_csv('./datasets/Tugas1.csv')
df_ytb_all
```

```python
df_ytb_all = pd.read_csv('./datasets/Tugas1.csv')
#df_ytb_all = df_ytb_all[(df_ytb_all['Source'] == '74.125.10.24') | (df_ytb_all['Destination'] == '74.125.10.24')]
df_ytb_all
```

```python
df_ytb_time = df_ytb_all[(df_ytb_all['Time'] < 1200.000000) & (df_ytb_all['Time'] > 600.00000 )]
df_ytb_time
```

```python
df_ytb_time = df_ytb_all[(df_ytb_all['Time'] < 1200.000000) & (df_ytb_all['Time'] > 600.00000 )]
df_ytb_tcp_udp = df_ytb_time[(df_ytb_all['Protocol'] == 'UDP') | (df_ytb_time['Protocol'] == 'TCP')]
df_ytb_tcp_udp = df_ytb_all[['Time', 'Length']]
df_ytb_tcp_udp
```

```python
df_ytb_udp = df_ytb_time[df_ytb_all['Protocol'] == 'UDP']
df_ytb_udp = df_ytb_all[['Time', 'Length']]
df_ytb_udp
```

```python
df1 = df_ytb_tcp_udp
```

```python
df2 = df_ytb_udp
```

```python
#membuat graph TCP_UDP
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(df1.index,df1["Length"])
plt.xlabel('index data')
plt.ylabel('packet length')
plt.show()
```

```python
#membuat graph UDP
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(df2.index,df2["Length"])
plt.xlabel('index data')
plt.ylabel('packet length')
plt.show()
```

```python
#TCP_UDP Plot Distribusi
plt.rcParams['figure.dpi'] = 100
sns.distplot(df1[df1['Length'].notnull()]['Length'], kde=False);
```

```python
plt.rcParams['figure.dpi'] = 100
sns.distplot(df2[df2['Length'].notnull()]['Length'], kde=False);
```

```python
#TCP_UDP Mean,Variance, SD
print('Mean TCP and UDP               : ', df1['Length'].mean())
print('Variance TCP and UDP           : ', df1['Length'].var())
print('Standard Deviation TCP and UDP : ', df1['Length'].std())
```

```python
#UDP Mean,Variance, SD
print('Mean UDP                       : ', df2['Length'].mean())
print('Variance UDP                   : ', df2['Length'].var())
print('Standard Deviation UDP         : ', df2['Length'].std())
```
