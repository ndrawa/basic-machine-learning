```python
df_ytb_all = pd.read_csv('./datasets/Tugas1.csv')
#df_ytb_all = df_ytb_all[(df_ytb_all['Source'] == '74.125.10.24') | (df_ytb_all['Destination'] == '74.125.10.24')]
df_ytb_all
```

```python
df_ytb_protocol = df_ytb_all[(df_ytb_all['Protocol'] == 'UDP') | (df_ytb_all['Protocol'] == 'TCP')]
df_ytb_protocol = df_ytb_protocol[['No.', 'Time', 'Length']]
```

```python
df_ytb_time = df_ytb_protocol[(df_ytb_protocol['Time'] < 1200.000000) & (df_ytb_protocol['Time'] > 600.00000 )]
df_ytb_time
```

```python
df_length = df_ytb_time['Length']
plt.plot(df_length)
```

```python
sns.distplot(df_ytb_time['Length'])
plt.show()
```

```python
sns.pairplot(data=df_ytb_time)
plt.show()
```

```python
df1 = pd.Series(df_ytb_time['Length']).value_counts().reset_index().sort_values('Length').reset_index(drop=True)
df1.columns = ['Length', 'Frequency']
dff = df1['Frequency']
df1[['Frequency','Length']].hist()

plt.show()
```
